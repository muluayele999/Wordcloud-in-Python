```python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator,WordCloud
from wordcloud import WordCloud
import os 
from os import path
os.chdir("C:\\Users\\rohan\\Desktop\\Virginia Tech\\Semester 3\\Library\\Dataset")
```

```python
# Read the whole text.
text1 = open("ea_telltale.txt").read()
text2 = open("ea_annabel.txt").read()
text3 = open("ea_raven.txt").read()
text4 = open("ea_thecask.txt").read()
text = text1 + text2 + text3+text4
text = text.replace('ugh','')
```

```python
#Upload mask for the wordcloud
d = "C:\\Users\\rohan\\Desktop\\Virginia Tech\\Semester 3\\Library\\Dataset\\examples"
mask = np.array(Image.open(path.join(d, "Raven_wing.jpg")))
```

```python
# word cloud creation
wordcloud = WordCloud(font_path="examples\\Garamond\\Garamond.ttf",max_font_size=200,background_color='black',width=100,height=50,
                     color_func=lambda *args, **kwargs: "white",mask=mask,
                     max_words=10000).generate(text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")

#set figure size and DPI
fig = plt.gcf()
fig.set_size_inches(76.8, 43.2,forward=True)
fig.savefig('py19.png', dpi=100, facecolor=fig.get_facecolor(), edgecolor='none')

plt.show()
```
